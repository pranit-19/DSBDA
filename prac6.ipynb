```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
data=pd.read_csv(r"C:\Users\dell\Downloads\Iris.csv")
```

```python
print(data.head())
```

```python
print("\nDataset Shape:",data.shape)
print("\nDataset Info:\n")
print(data.info())
print("\nChecking Missing Values:\n")
print(data.isnull().sum())
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
print("\n Feature Data(X):\n",X.head())
print("\n Target Data(y):\n",y.head())
```

```python
X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.3,random_state=42
)

print("\nNumber of Training Samples:",len(X_train))
print("Number of Testing Samples:",len(X_test))
```

```python
model=GaussianNB()


model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
print("\n Model Training Completed Successfully!")
print("\n Actual Values:\n",list(y_test))
print("\nPredicted Values:\n",list(y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
print("Confusion Matrix:\n",cm)
```

```python
labels = sorted(y_test.unique())
cm_df = pd.DataFrame(cm, index=labels, columns=labels)
print("\nLabeled Confusion Matrix:\n")
print(cm_df)
```

```python
overall_accuracy=accuracy_score(y_test,y_pred)
print("\nOverall Accuracy:",overall_accuracy)
```

```python
train_acc=model.score(X_train,y_train)
test_acc=model.score(X_test,y_test)

print("\nTraining Accuracy=",train_acc)
print("Testing Accuracy=",test_acc)
```

```python
print("\nClassification Report:\n")
print(classification_report(y_test,y_pred))
```

```python
print("Calculation of TP, FP, FN, TN for each class:\n")
total=cm.sum()
for i in range(len(cm)):
    TP=cm[i][i]
    FP=cm[:,i].sum()-TP
    FN=cm[i,:].sum()-TP
    TN=total-(TP+FP+FN)
print("For Class",i)
print("TP=",TP)
print("FP=",FP)
print("FN=",FN)
print("TN=",TN)
    
    
    
```

```python
TP=cm[0][0]
FP=cm[:,0].sum()-TP
FN=cm[0,:].sum()-TP
TN=total-(TP+FP+FN)
print("TP=",TP)
print("FP=",FP)
print("FN=",FN)
print("TN=",TN)
```

```python
accuracy=(TP+TN)/(TP+TN+FP+FN)
error_rate=1-accuracy
precision=TP/(TP+FP)
recall=TP/(TP+FN)

print("Accuracy:",accuracy)
print("Error Rate:",error_rate)
print("Precision:",precision)
print("Recall",recall)
```
