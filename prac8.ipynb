```python
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
titanic=pd.read_csv(r"C:\Users\dell\Downloads\tested.csv")
```

```python
print(titanic.head())
```

```python
print(titanic.info())
```

```python
#check survival count 
sns.countplot(x="Survived",data=titanic)
plt.title("Survival Count(0=No,1=Yes)")
plt.show()
```

```python
#survival Count based on gender
sns.countplot(x="Survived",hue="Sex",data=titanic)
plt.title("Survival by gender")
plt.show()
```

```python
#survival based on Passenger class
sns.countplot(x="Survived",hue="Pclass",data=titanic)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
#Age Distribution
sns.histplot(titanic["Age"].dropna(),bins=30,kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
#Survival Vs Age
sns.boxplot(x="Survived",y="Age",data=titanic)
plt.title("Age vs Survival")
plt.show()
```

```python
#Fare Distribution by class 
sns.boxplot(x="Pclass",y="Fare",data=titanic)
plt.title("Fare Distribution by Class")
plt.show()
```

```python
#HeatMap(Correalation)
corr=titanic.corr(numeric_only=True)
```

```python
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#write a code to check how the price of the ticket (column name:"fare") is distributed by plotting a histogram

#create histogram for "fare"
plt.figure(figsize=(8,5))
sns.histplot(titanic["Fare"],bins=30,kde=True)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```
